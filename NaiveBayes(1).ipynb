```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
es=pd.read_csv('/content/enjoysports.csv')
es.head()
```

```python
le = LabelEncoder()
for column in es.columns[:-1]:es[column]=le.fit_transform(es[column])

M=es.iloc[:,:-1].values
n=es.iloc[:,-1].values
```

```python
M_train,M_test,n_train,n_test=train_test_split(M,n,test_size=0.2,random_state=42)
```

```python
mdl=GaussianNB()

mdl.fit(M_train,n_train)
n_pred=mdl.predict(M_test)
```

```python
cm=confusion_matrix(n_test,n_pred)
accuracy=accuracy_score(n_test,n_pred)
report=classification_report(n_test,n_pred)

print("Confusion Matrix caluclated is:\n", cm)
print("Accuracy is:", accuracy)
print("Classification Report is:\n", report)
```

```python
error_rate =1-accuracy
print("Error Rate is:",error_rate)
```
